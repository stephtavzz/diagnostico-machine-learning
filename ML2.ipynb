##**Escreva uma função que receba uma lista de números e retorne outra lista com os números ímpares.**

```python
def numeros_impares():
    lista1 = [24, 35, 9, 6, 42, 59, 63, 10, 75, 88, 91]
    lista2 = []

    for numero in lista1:
        if numero % 2 == 1:  # Verifica se o número é ímpar
            lista2.append(numero)

    return lista2

lista = numeros_impares()
lista
```

## **Escreva uma função que receba uma lista de números e retorne outra lista com os números primos presentes.**

```python
def numeros_primos():
    lista1 = [24, 35, 9, 6, 42, 59, 63, 10, 75, 88, 91]
    lista2 = []

    for numero in lista1:
        if numero > 1:
            for i in range(2, int(numero ** 0.5) + 1):
                if numero % i == 0:
                    break
            else:
                lista2.append(numero)

    return lista2

lista = numeros_primos()
lista
```

##**Escreva uma função que receba duas listas e retorne outra lista com os elementos que estão presentes em apenas uma das listas.**


```python
def elementos_distintos():
    lista1 = ['Maçã', 1.80, 'Paraguai', True, 15]
    lista2 = ['Melancia', 1.60, 'Paraguai', False, 15]

    # Utilizando a diferença simétrica para encontrar elementos únicos
    resultado = list(set(lista1) ^ set(lista2))

    return resultado

lista_resultante = elementos_distintos()
lista_resultante
```

##**Dada uma lista de números inteiros, escreva uma função para encontrar o segundo maior valor na lista.**




```python
def segundo_maior_valor():
    lista = [400, 89, 550, 321, 240, 930, 36]

    maior_valor = max(lista)  # Encontrando o maior valor
    lista.remove(maior_valor)  # Removendo o maior valor da lista

    segundo_valor = max(lista)  # Encontrando o segundo maior valor

    return segundo_valor

segundo_maior = segundo_maior_valor()
segundo_maior
```

## **Crie uma função que receba uma lista de tuplas, cada uma contendo o nome e a idade de uma pessoa, e retorne a lista ordenada pelo nome das pessoas em ordem alfabética.**

```python
def nome_ordenado():
    lista_tuplas = [('Ana', 40), ('Renata', 35), ('Rafael', 45), ('Marcos', 50), ('Amanda', 26)]

    # Ordenando a lista de tuplas pelo primeiro elemento (nome)
    lista_tuplas.sort(key=lambda x: x[0])

    return lista_tuplas

lista_ordenada = nome_ordenado()
lista_ordenada
```

##**Como identificar e tratar outliers em uma coluna numérica usando desvio padrão ou quartis?**

**Outliers** em colunas numéricas podem ser identificados usando **desvio padrão** ou **quartis (IQR)**. Com o desvio padrão, define-se um **intervalo** em torno da **média** (como **média ± 2 ou 3 desvios**, dependendo do **nível** de rigor desejado), e *remove-se* o que estiver fora desse intervalo. Já com o **IQR**, usa-se os **quartis** para detectar valores extremos. Ambos os métodos ajudam a limpar os dados e evitar distorções em análises.

```python
import pandas as pd

dados = pd.DataFrame({'salario': [3000, 3000, 3000, 3000, 3000, 6000]})

# Cálculo da média e desvio padrão
media = dados['salario'].mean()
desvio = dados['salario'].std()

# Limites de outliers (2 * desvio padrão)
limite_inferior = media - 2 * desvio
limite_superior = media + 2 * desvio

# Filtra os dados sem outliers
dados_sem_outliers = dados[(dados['salario'] >= limite_inferior) & (dados['salario'] <= limite_superior)]

print("Média:", media)
print("Desvio padrão:", desvio)
print("Limite inferior:", limite_inferior)
print("Limite superior:", limite_superior)
print("\nDados sem outliers:")
dados_sem_outliers
```

##**Como concatenar vários DataFrames (empilhando linhas ou colunas), mesmo que tenham colunas diferentes?**

### **Dica:** Utiliza-se *pd.concat()* especificando *axis=0 (linhas)* ou *axis=1(colunas)*. Quando há colunas diferentes, os valores ausentes são preenchidos com **NaN**.


```python
import pandas as pd

df1 = pd.DataFrame({'nome': ['Ana', 'João'], 'idade': [25, 30]})
df2 = pd.DataFrame({'nome': ['Carlos'], 'salario': [3000]})

df_linhas = pd.concat([df1, df2], axis=0)

df_linhas
```

##**Utilizando pandas, como lidar com valores ausentes (NaN) em um DataFrame?**


Para lidar com valores ausentes em um DataFrame com pandas, utiliza-se **df.isnull().sum()** para identificar a quantidade de NaNs por coluna. Depois, escolhe-se uma estratégia de imputação: para variáveis numéricas, pode-se usar 0, **mean()**, **median()** ou **mode()**; para categóricas, **mode()** ou uma nova categoria como "Desconhecido". Se a imputação não for adequada, é possível remover linhas ou colunas com **df.dropna()**, sempre avaliando o impacto nos dados.
